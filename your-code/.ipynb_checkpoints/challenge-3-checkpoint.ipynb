#### Use the appropriate visulalization to show the distribution of the column `Age`

```python
# your code here
titanic.groupby(['Age'])['PassengerId'].count().plot(kind='hist',figsize=(12,6),bins=20)
```

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
fig,[ax1,ax2,ax3] = plt.subplots(1,3,figsize=(12,6))

ax1.hist(titanic.groupby(['Age'])['PassengerId'].count(),bins=10)
ax2.hist(titanic.groupby(['Age'])['PassengerId'].count(),bins=20)
ax3.hist(titanic.groupby(['Age'])['PassengerId'].count(),bins=50)

ax1.set_title('bin=10')
ax2.set_title('bin=20')
ax3.set_title('bin=50')
plt.show()
```
